```python
%%sql

SELECT
	DISTINCT Country_name,
	Time_period,
	Est_population_in_millions
FROM
	united_nations.Access_to_Basic_Services
LIMIT 5;
```

## 1. Combine the columns

```python
%%sql

SELECT
	DISTINCT Country_name,
	Time_period,
	Est_population_in_millions,
    CONCAT(
        Country_name, 
        Time_period, 
        Est_population_in_millions
    ) as Country_id
FROM
	united_nations.Access_to_Basic_Services
LIMIT 5;
```

## 2. Replace the NULL values

```python
%%sql

SELECT
	DISTINCT Country_name,
	Time_period,
	Est_population_in_millions,
    CONCAT(
        ifnull(Country_name, "UNKNOWN"), 
        ifnull(Time_period, "UNKNOWN"), 
        ifnull(Est_population_in_millions, "UNKNOWN")
    ) as Country_id
FROM
	united_nations.Access_to_Basic_Services
LIMIT 10;
```

## 3a. Make the custom IDs consistent

```python
%%sql

SELECT
	DISTINCT Country_name,
	Time_period,
	Est_population_in_millions,
    CONCAT(
        LEFT(UPPER(IFNULL(Country_name, "UNKNOWN")), 4), 
        LEFT(UPPER(IFNULL(Time_period, "UNKNOWN")), 4), 
        RIGHT(UPPER(IFNULL(Est_population_in_millions, "UNKNOWN")), 7)
    ) as Country_id
FROM
	united_nations.Access_to_Basic_Services
LIMIT 10;
```

```python
%%sql

SELECT
	DISTINCT Country_name,
	Time_period,
	Est_population_in_millions,
	CONCAT(
		SUBSTRING(IFNULL(UPPER(Country_name), 'UNKNOWN'),1,4),
		SUBSTRING(IFNULL(Time_period, 'UNKNOWN'),1,4),
		SUBSTRING(IFNULL(Est_population_in_millions, 'UNKNOWN'),-7)
	) AS Country_id
FROM
	united_nations.Access_to_Basic_Services;
```
