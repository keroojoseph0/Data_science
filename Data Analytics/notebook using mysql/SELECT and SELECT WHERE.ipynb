```python
%%sql

SELECT
    country_name,
    time_period,
    pct_managed_drinking_water_services AS pct_access_to_water
FROM
    united_nations.Access_to_Basic_Services
WHERE
    Time_period = 2020
LIMIT 100;
```

```python
%%sql

SELECT
    country_name,
    time_period,
    pct_managed_drinking_water_services AS pct_access_to_water
FROM
    united_nations.Access_to_Basic_Services
WHERE
    Time_period = 2020
ORDER BY pct_access_to_water
LIMIT 10;
```
